```python
#vector embeddings and storing in chromadb
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma

embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001", task_type="RETRIEVAL_QUERY")

db = Chroma.from_documents(document[:5], embeddings)
```

```python
query = "what does the page say about llm"
result = db.similarity_search(query)
result
```

```python
from langchain_google_genai import GoogleGenerativeAI
llm = GoogleGenerativeAI(model="gemini-3-flash-preview")
llm
```

```python
from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate.from_template("""answer the following question based only on the provided context think step br step before providing a detailed answer
<context>{context}</context>
question: {input} """)
```

```python
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
chain = create_stuff_documents_chain(llm , prompt)
```

```python
reteriver = db.as_retriever()
reteriver
```

```python
from langchain_classic.chains import create_retrieval_chain
reterival_chain = create_retrieval_chain(reteriver , chain )
reterival_chain.invoke({"input":"scaled dot product attention"})
```
