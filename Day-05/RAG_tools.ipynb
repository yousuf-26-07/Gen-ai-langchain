```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://www.apple.com/in/iphone/")
web_doc=loader.load()
web_doc
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
document = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=200).split_documents(web_doc)
document
```
